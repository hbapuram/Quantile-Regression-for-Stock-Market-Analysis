```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
infosys = pd.read_csv('INFY.csv')
infosys.head()
infosys.info()
```

```python
INFY = []
for i in range(1,2863):
    s = round(((infosys['Close'][i]-infosys['Close'][i-1])/(infosys['Close'][i-1]))*100, 4)
    INFY.append(s)
INFY
Date = pd.to_datetime(list(infosys['Date']))
```

```python
INFY = pd.DataFrame(list(zip(Date, INFY)), columns = ['Date', 'INFY'])
INFY.head()
```

```python
plt.figure()
plt.plot(INFY['Date'], INFY['INFY'], color = 'r', label = 'INFY')
plt.legend()
plt.show()
```

```python
from functools import reduce
stocks = reduce(lambda x,y: pd.merge(x,y, on='Date', how='outer'), [INFY, TCS, AMZN, GOOG, MSFT, HTCY])
```

```python
stocks.info()
```

```python
st = stocks.copy()
st
```

```python
st.isna().sum()
```

```python
st["INFY"] = pd.to_numeric(st.INFY, errors='coerce')
st["TCS"] = pd.to_numeric(st.TCS, errors='coerce')
st["AMZN"] = pd.to_numeric(st.AMZN, errors='coerce')
st["GOOG"] = pd.to_numeric(st.GOOG, errors='coerce')
st["MSFT"] = pd.to_numeric(st.MSFT, errors='coerce')
st["HTCY"] = pd.to_numeric(st.HTCY, errors='coerce')
```

```python
st_nonmiss = st.dropna()
med_INFY, med_TCS, med_AMZN, med_GOOG, med_MSFT, med_HTCY = np.median(st_nonmiss['INFY']), np.median(st_nonmiss['TCS']), np.median(st_nonmiss['AMZN']), np.median(st_nonmiss['GOOG']), np.median(st_nonmiss['MSFT']), np.median(st_nonmiss['HTCY'])
```

```python
st['INFY'].fillna(value =  med_INFY, inplace = True)
st['TCS'].fillna(value =  med_TCS, inplace = True)
st['AMZN'].fillna(value =  med_AMZN, inplace = True)
st['GOOG'].fillna(value =  med_GOOG, inplace = True)
st['MSFT'].fillna(value =  med_MSFT, inplace = True)
st['HTCY'].fillna(value =  med_HTCY, inplace = True)
st.head()
```

```python
def log_transform(x):
    if x > 0:
        x = np.log(x)
    else:
        x = -np.log(x)
```

```python
st['INFY'] = st['INFY'].apply(lambda x : np.log(x) if x>0 else (-np.log(x)))
st['TCS'] = st['TCS'].apply(lambda x : np.log(x) if x>0 else (-np.log(x)))
st['AMZN'] = st['AMZN'].apply(lambda x : np.log(x) if x>0 else (-np.log(x)))
st['GOOG'] = st['GOOG'].apply(lambda x : np.log(x) if x>0 else (-np.log(x)))
st['MSFT'] = st['MSFT'].apply(lambda x : np.log(x) if x>0 else (-np.log(x)))
st['HTCY'] = st['HTCY'].apply(lambda x : np.log(x) if x>0 else (-np.log(x)))
```

```python
fig, axis = plt.subplots(2,3, figsize = (20,10))
sns.kdeplot(st['INFY'], ax = axis[0,0])
sns.kdeplot(st['TCS'], ax = axis[0,1], color = 'r')
sns.kdeplot(st['AMZN'], ax = axis[0,2], color = 'y')
sns.kdeplot(st['GOOG'], ax = axis[1,0], color = 'g')
sns.kdeplot(st['MSFT'], ax = axis[1,1], color = 'orange')
sns.kdeplot(st['HTCY'], ax = axis[1,2], color = 'purple')
```

```python
plt.figure(figsize = (10,5))
sns.heatmap(st.corr(), cmap = 'BuGn', annot = True)
```

```python
fig, axes = plt.subplots(2,2, figsize = (8,8))
sns.scatterplot(st['TCS'], st['AMZN'], color = 'r', ax = axes[0,0])
axes[0,0].set_title('TCS VS AMZN')
sns.scatterplot(st['TCS'], st['GOOG'], color = 'orange', ax = axes[0,1])
axes[0,1].set_title('TCS VS GOOG')
sns.scatterplot(st['TCS'], st['MSFT'], color = 'green', ax = axes[1,0])
axes[1,0].set_title('TCS VS MSFT')
sns.scatterplot(st['TCS'], st['HTCY'], color = 'blue', ax = axes[1,1])
axes[1,1].set_title('TCS VS HTCY')
```
